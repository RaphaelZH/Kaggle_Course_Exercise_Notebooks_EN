```python
def environnement(x, y, var_x_y):
    reward     = 0
    new_x      = x + var_x_y[0]
    new_y      = y + var_x_y[1]

    if new_x<0 : new_x = 0
    if new_x>3 : new_x = 3
        
    if new_y<0 : new_y = 0
    if new_y>2 : new_y = 2
    
    if new_x ==2 and new_y==1 : reward=-10
    if new_x ==3 and new_y==2 : reward=+50
    
    est_fini = reward != 0
    
    return new_x, new_y, reward, est_fini
```
