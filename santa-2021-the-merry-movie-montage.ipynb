# Permutation

```python
import itertools

print("Elements' permutations: {}\nNumber of permutations: {}\n"
      .format(list(itertools.permutations([1, 2, 3])), len(list(itertools.permutations([1, 2, 3])))))
print("Elements' permutations: {}\nNumber of permutations: {}\n"
      .format(list(itertools.permutations([1, 2, 3], 2)), len(list(itertools.permutations([1, 2, 3], 2)))))
print("Elements' permutations: {}\nNumber of permutations: {}"
      .format(list(itertools.permutations([1, 2, 3], 1)), len(list(itertools.permutations([1, 2, 3], 1)))))
```

```python
comm =permutations[permutations.Permutation.str[:2] =='ðŸŽ…ðŸ¤¶']
comm
```

```python
other =permutations[permutations.Permutation.str[:2] !='ðŸŽ…ðŸ¤¶'].reset_index(drop=True)
other
```

```python
a=comm.copy()
b=comm.copy()
c=comm.copy()
for i in range(len(other)):
    if i%3 ==0:
        a = a.append(other.iloc[i],ignore_index=True)
    if i%3 ==1:
        b = b.append(other.iloc[i],ignore_index=True)
    if i%3 ==2:
        c = c.append(other.iloc[i],ignore_index=True)
        
aa =str()
bb =str()
cc =str()
for i in a.Permutation.values:
    aa +=i

for i in b.Permutation.values:
    bb +=i
for i in c.Permutation.values:
    cc +=i
    
    
sample_submission.loc[0,'schedule'] = aa
sample_submission.loc[1,'schedule'] = bb
sample_submission.loc[2,'schedule'] = cc
```

```python
print(len(sample_submission.loc[0,'schedule']))
print(len(sample_submission.loc[1,'schedule']))
print(len(sample_submission.loc[2,'schedule']))
```
