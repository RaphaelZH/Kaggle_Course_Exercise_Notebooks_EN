# Permutation

```python
import itertools

print("Elements' permutations: {}\nNumber of permutations: {}\n"
      .format(list(itertools.permutations([1, 2, 3])), len(list(itertools.permutations([1, 2, 3])))))
print("Elements' permutations: {}\nNumber of permutations: {}\n"
      .format(list(itertools.permutations([1, 2, 3], 2)), len(list(itertools.permutations([1, 2, 3], 2)))))
print("Elements' permutations: {}\nNumber of permutations: {}"
      .format(list(itertools.permutations([1, 2, 3], 1)), len(list(itertools.permutations([1, 2, 3], 1)))))
```
